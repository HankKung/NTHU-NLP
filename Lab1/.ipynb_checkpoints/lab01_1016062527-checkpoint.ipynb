```python
import os
import sys
import pandas as pd
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
import math
from collections import Counter, defaultdict
d = 0.75
```

```python
def read_data(PATH):
    
    all_data = []
    for path, dirs, files in os.walk(PATH):
        temp = []
        for filename in files:
            filePath = path + '/' + filename
            #print(filePath)
            f = open( filePath, 'r').read()
            temp.append(parse_article(f))

        if temp:
            all_data.append(temp)
    return all_data
```

```python
def parse_article(f):
#     TODO: tokenize sentence and transform to lower and padding
    article = sent_tokenize(f)
    for sen in article:
        sen = sen.lower()
    return article
```

```python
def kn_model(bi ,label, str_1, str_2):
#     TODO: Bigram() + Pcontination()
# KN( wi | wi−1 ) = log( Bigram + λ(wi−1)* PCONTINUATION)
# Bigram : max( C(wi-1, wi) -d )/∑C(wi−1, w*), 0 )
# PCONTINUATION: max( ∑type(w* ,wi)- d )/∑type(wj-1 ,wj) , 0 )
# λ(wi−1) = (d / ∑C(wi−1, w*))*∑type(wi−1 ,w*)
    bigram_word = str_1 +' '+str_2
    bigram_word_num = bi[label][bigram_word]
    temp_str1 = str_1
    temp_len = len(temp_str1)
    prefix = 0.0000001
    prefix__type_num = 0
    for key in bi[label]:
        if key[0:temp_len] == temp_str1:
            prefix += bi[label][key]
            prefix__type_num += 1
    
    bigram = max(bigram_word_num, 0)/prefix
    
    temp_str2 = str_2
    temp_len2 = len(temp_str2)
    type_of_postfix = 0
    for key in bi[label]:
        if key[-temp_len2:] == temp_str2:
            type_of_postfix+=1
#     print(bigram)
#     print(prefix)
#     print(prefix__type_num)
    p_con = max((type_of_postfix-d)/len(bi[label]), 0)
    
    lam = (d/prefix)*prefix__type_num
#     print(p_con)
#     print(lam)
    kn = math.log(bigram + (lam*p_con))
    return kn
```

```python
def padding(l):
    l.insert(0,'<s>')
    l.append("<\s>")
    return l
```

```python
def counter(l):
    return Counter(l)
```

```python
def count_2gram(l):
    temp = []
    for idx, val in enumerate(l):
        if idx+1 < len(l):
            temp.append(val +' '+ l[idx+1])
    return Counter(temp)
```

```python
def progress(count, total, hit, total_c, status=''):
    t = hit/total_c
    bar_len = 60
    filled_len = int(round(bar_len * count / float(total)))

    percents = round(100.0 * count / float(total), 1)
    bar = '=' * filled_len + '-' * (bar_len - filled_len)

    sys.stdout.write('[%s] %s%s ...%s %s\r' % (bar, percents, '%', status, t))
    sys.stdout.flush()
```
